```python
#DICTIONARY TO STORE THE MAIN MENU's MEAL OPTIONS AND PRICES
menu_options = {"ewedu": 100, "gbegiri": 100, "ilalasepo": 150, "egusi": 150, "efo riro": 200,
                "edika": 200, "afang": 200, "fisherman": 200, "eba": 100, "semo": 100, "amala" : 100, "poundo": 200, 
                "white rice": 100, "jollof rice": 200, "coconut rice": 200, "traditional rice": 200, "fried rice": 300,
                "beef": 100, "titus fish": 100, "orisirisi": 100, "chicken": 300, "croaker": 300, "catfish": 400, 
                "plantain": 100, "beans": 100, "moimoi": 100, "premium moimoi": 200, "agege bread": 200,
                "sliced bread": 200, "coleslaw": 200, "boiled egg": 100,
                "satchet water": 20,"bottle water": 100, "coke": 100, "fanta": 100, "sprite": 100, "malt": 200,"fura": 300}

#DICTIONARY TO STORE THE SPECIAL MENU's MEAL OPTIONS AND PRICES
#Starts empty for manager to update as desired.
special_options = {}

def main_menu():
    """This function is to print the meal options on the main menu if it's populated.
    It will trigger a comment if the main menu is empty."""
    if menu_options != {}:
        print("The main menu options:")
        count = 0
        for k in menu_options:
            count +=1
            print("MEAL #:", [count], k.upper(), "is sold at", menu_options[k], "NGN.")
        choose_main()  
    else:
        print("This menu is currently unavailable. Pls try our other menu option(s).")

            
def specials_menu():
    """This function is to print the meal options on the specials menu if it's populated.
    It will trigger a comment if the specials menu is empty."""
    if special_options != {}:
        print("The specials for today are:")
        count = 0
        for k in special_options:
            count +=1
            print("SPEC #:", [count], k.upper(), "is sold at", special_options[k], "NGN.")
        choose_specials()
    else:
        print("This menu is currently unavailable. Pls try our other menu option(s).")     
    
def menu_pick():
    """This function gives selection options for the menu options, takes the user input and returns it as a integer.
    It gives a message for non-integer input."""
    try:
        choice = input("Please enter:\n[1]MAIN MENU\n[2]SPECIALS\n[3]To Return to the General Operations Menu.\nYour Entry is:")
        return int(choice)
    except ValueError:
        print("Process selection error.\nOnly 1, 2 or 3 is valid. Please check your entry and retry.")
        
def see_menu():
    """This function triggers other funtion(s) to display the selected menu option from the menu_pick(). 
    It gives a comment for invalid entry."""
    print("You can make selection(s) from our Main Menu or Specials:")
    while True:
        next = menu_pick()
        if next == 1:
            main_menu()
            continue
        elif next == 2:
            specials_menu()
            continue
        elif next == 3:
            print("Returning to the General Operations Menu.")       
            break                        #To exit the menu selections page
        else:
            print("Invalid entry. Only 1 and 2 are valid. Please check your entry and retry.")
            continue                        

#CREATING EMPTY LISTS TO STORE CUSTOMER ORDER AND ORDER FEATURES
meals = []
rates = []
qtys = []
costs = []

#THE NEXT SET OF FUNCTIONS ARE TO TAKE THE CUSTOMER ORDER
def choose_meal():
    """This function gives instruction to choose a meal from the selected menu, takes the user input and returns it as an integer.
    It gives a message for non-integer input."""
    try:
        choice = input("""\nTo select your order from the current menu option, enter the MEAL # or SPEC #.
To return to the General Operations Menu, enter 0 : """)
        return int(choice)
    except ValueError:
        print("Selection error.\nOnly meal or special numbers from the menu options are valid. Please check your entry and retry.")

def order_qty():
    """This function gives instruction for a user to provide quantity for the selected meal, takes the user input and returns it as a integer.
    It gives a message for non-integer input."""
    try:
        qty = input("How many of that would you be having?: ")
        return int(qty)
    except ValueError:
         print("Quantity input error.\nYou can enter only numbers. Please check your entry and retry.")
                

#FUNCTION TO CHOOSE FROM THE MAIN MENU
def choose_main():
    """This function handles user selection from the mainmenu.
    It ensures the selection and is within the meal range and stores valid selctions with their qty, rate and calculated cost.
    It gives an output for invalid input."""
    max_choice = len(menu_options)
    min_choice = 1
    while True:
        choice = choose_meal()
        if choice == None:
            continue
        elif choice == 0:
            print("Ok. That's all for now.")
            break
        elif not min_choice <= choice <= max_choice:
            print("That's not a valid meal option.")
            continue
        elif min_choice <= choice <= max_choice:          
            while True:
                qty = order_qty()
                if qty != None:
                    menu_choice = list(menu_options.keys())[choice - 1]
                    rate = list(menu_options.values())[choice - 1]
                    cost = qty * rate
                    meals.append(menu_choice)
                    qtys.append(qty)
                    rates.append(rate)
                    costs.append(cost)
                    break
                else:
                    continue
        else:
            print("Input error.")
            break

#FUNCTION TO CHOOSE FROM SPECIALS MENU
def choose_specials():
    """This function handles user selection from the specials menu.
    It ensures the selection and is within the meal range and stores valid selctions with their qty, rate and calculated cost.
    It gives an output for invalid input."""
    max_choice = len(special_options)
    min_choice = 1
    while True:
        choice = choose_meal()
        if choice == None:
            continue
        elif choice == 0:
            print("Ok. That's all for now.")
            break
        elif not min_choice <= choice <= max_choice:
            print("That's not a valid specials option.")
            continue
        elif min_choice <= choice <= max_choice:          
            while True:
                qty = order_qty()
                if qty != None:
                    menu_choice = list(special_options.keys())[choice - 1]  
                    rate = list(special_options.values())[choice - 1]
                    cost = qty * rate
                    meals.append(menu_choice)
                    qtys.append(qty)
                    rates.append(rate)
                    costs.append(cost)
                    break
                else:
                    continue
        else:
            print("Input error.")

def so_far():
    """This function prints the available customer order.
    It gives a message for an empty order."""
    if meals != []:
        count = 0
        print("[S/N]\tMEAL NAME\tQTY\tRATE\tCOST")
        for meal, qty, rate, cost in zip(meals, qtys, rates, costs):
            count +=1
            print([count],"   ", meal.upper(),"      ", qty,"   ", rate,"   ", cost,"NGN")
        print(f"\nOrder total(before Tax): {sum(costs):,.2f} NGN")
        print("Sales tax: 10%")
        print(f"TOTAL DUE: {tax(sum(costs)):,.2f} NGN\n")
    else:
        print("**You have no order on record.")    
               
def tax(x):
    """Function to apply tax on the order cost"""
    rate = 1.10
    return x *rate

#IMPORT DATETIME MODULE FOR A FORMATTED DATE VARIABLE 
from datetime import *
today = date.today()
date = "{0:%A %B %d, %Y.}".format(today)

#FUNCTION TO OUTPUT RECEIPT BASED ON THE CUSTOMER'S ORDER
def receipt():
    name = input("Please Enter customer's name for receipt: ").upper()
    print("-"*65)
    print("*"*28, "RECEIPT", "*"*28)
    print(date)
    print("HELLO", name,"!")
    print("Here's your completed order from OKO IYA'LAMALA KITCHEN")
    so_far()
    print("\nMeals are best consumed within 30 mintues from purchase.")
    print("THANKS FOR YOUR PATRONAGE! DO CALL AGAIN.")
    print("*"*65,"\nFor inquiry/complaints:\nCall: +(234)OKOIYALAMALA\nE-mail: customer@okoiyalamala.ng")
    print("Follow or Tag us Twitter and IG: @okoiyalamala.ng")
    print("-"*65)
    return
 
#MANAGER MODE
#SECURITY TO ENSURE ONLY AUTHORIZED USERS CAN ACCESS MANAGER FEATURES
from getpass import getpass
def user():
    """This function takes and tests the username and passcode for access to the manager mode.
    It works (indirectly) with two other functions."""
    while True:
        name = input("\nThis is the Manager Menu.\nEnter username or # to exit: ").upper()
        if name == "MANAGER":
            return test_pass()             
        elif name == "#":
            return "Ok. Bye!"
        else:
            print("Wrong username.")
    return    

def check_manager():
    "This function uses getpass to collect and hide characters for the manager's password."
    print('Okay MANAGER, enter the correct password to proceed.\nPress # to exit')
    return getpass("Password: ")
      
def test_pass():
    """This function checks for the correct manager password within 5 attempts.
    It denies access when the trials exceed 5."""
    trials = 1  
    while trials < 6:
        passcode = check_manager()        
        if passcode == "shalom":
            print("WELCOME BOSS! You're in charge.")
            manager()
        elif passcode == "#":
            print("Exiting Manager Mode.")
            break
        else:
            print('Invalid Entry. Retry!')
            print(f"{5- trials} attempt(s) left\n")
        trials +=1
        
    if trials == 6:
        print("You've reached the attempt limit. You can't access the Manager Mode.")

         
#TO MODIFY A CUSTOMER ORDER ENTRY
def ask():
    """This function gives instructions for the operation options, takes the user input and returns it as an integer.
    It gives a message for non-integer input."""
    try:
        ask =input("Enter:\n1 To edit an order entry.\n2 To Exit.")
        return int(ask)
    except ValueError:
        print("You can only enter numbers.")
        
def change():
    if meals != []:
        while True:
            option = ask()
            if option == 1:
                print("***********Here's the available order***********")
                so_far()          #Not needed for run since it will already be printed
                while True:
                    modify = input("""\nWhat order entry do you want delete? Please enter the meal name correctly.\nOr press Enter to exit.
                    Your entry:""").lower()                        # Ensuring lowercase input to deactivate case sensitivity
                    if modify != "" and modify in meals:
                        print("Do you want to remove", modify.upper(), "from your order?")
                        while True:
                            confirm = input("Please enter yes or no.\nYour entry: ").lower()                        # Ensuring lowercase input to deactivate case sensitivity
                            if confirm == "yes":                                                       # Condition for a confirmed update 
                                x = meals.index(modify)
                                meals.remove(modify)
                                del rates[x]
                                del qtys[x]
                                del costs[x]
                                print("**", modify.upper(),"deleted.\nNow your order is:")
                                so_far()
                                break 
                            elif confirm == "no":                                                                # Condition for a no input
                                print("Okay, request cancelled!", modify.upper(), "was not removed from the order.")        # An output response   
                                break                                                                           # End the program by breaking the while loop
                            elif confirm == "":
                                print("No input given.")
                            else:                                                                   # Condition for an input that is neither yes nor no
                                print("Invalid Entry.")
                                break

                    elif modify == "":
                        print("Ok, no changes made.")
                        break
                    else:
                        print("""Invalid Entry. Only a meal name on the exisiting order is allowed.
                If you entered a meal name, check that you entered it correctly and RETRY!""")
                break 
            elif option == 2:
                print("No Changes made.")
                break
            else:
                print('Invalid Entry. Retry!')
    else:
        print("There's no order to change.")
        
               
#FUNCTION ADD TO THE MAIN_MENU
def price():
    """This function gives instructions for the manager to input price, takes the input and returns it as an integer.
    It gives a message for non-integer input."""
    try:
        value = input("How much is it?:")
        return int(value)
    except ValueError:
        print("You can only enter numbers for the price. Please check your entry and retry.")
        
def main_update():
    newmeal = {input("Enter food name:"): price()}            
    for k in newmeal:
        while k != None and newmeal[k] != None:                
            print("Do you want to add", k.upper(), "worth", newmeal[k],"NGN as a new entry on the main menu?")      #Confirming request
            confirm = input("Please enter yes or no.\nYour entry: ").lower()                            # Ensuring lowercase input to deactivate case sensitivity
            if confirm == "yes":                                                                 # Condition for a confirmed request 
                menu_options.update(newmeal)                                                # Variable to confirm a random selection from fortunes
                print("**Done. Return to the general operations menu to see the updated menu.")
                break                                                                          # An output of the random selection
            elif confirm == "no":                                                              # Condition for a no input
                print("Okay, request cancelled! Nothing was added to the main menu.")          # An output response   
                break                                                                          # Cancellation of the process by breaking the while loop
            elif confirm != 'yes' or 'no':                                                     # Condition for an invalid input
                print("Invalid entry.")
        else:
            print("Invalid entry for the name and/or price. Start over.")
            break    

#FUNCTION DELETE FROM THE MAIN_MENU
def del_main_menu():
    while menu_options != {}:                                              #Condition for an unempty main menu
        modify = input("""\nWhat option do you want delete? Please enter the meal name correctly.\nOr press Enter to exit.
        Your entry:""").lower()                        # Ensuring lowercase input to deactivate case sensitivity
        if modify != "" and modify in menu_options:                              #Condition for given input that's on the main menu
            print("Do you want to remove", modify.upper(), "from the menu options?")
            while True:
                confirm = input("Please enter yes or no.\nYour entry: ").lower()                            # Ensuring lowercase input to deactivate case sensitivity
                if confirm == "yes":                                                       # Condition for a confirmed update 
                    menu_options.pop(modify)
                    print("**Entry deleted. Return to the general operations menu to see the update.")
                    break 
                elif confirm == "no":                                                                # Condition for a no input
                    print("Okay, request cancelled!", modify.upper(), "was not removed from the menu options")        # An output response   
                    break                                                                           # End the program by breaking the while loop
                elif confirm == "":
                    print("No input given.")
                else:                                                                   # Condition for an input that is neither yes nor no
                    print("Invalid Entry.")
                    break
        elif modify == "":
            print("Ok, no changes made.")
            break
        else:
            print("""Invalid Entry. Only a meal name on the menu is allowed.
    If you entered a meal name, check that you entered it correctly then retry.""")
    if menu_options == {}:                                                       #Condition for an empty main menu
        print("This menu is currently unavailable. No changes can be made.")


#FUNCTION ADD TO THE SPECIALS_MENU
def specials_update():                                            
    spec = {input("Enter food name:"): price()}
    for k in spec:
        while k != "" and spec[k] != None:                
            print("Do you want to add", k.upper(), "worth", spec[k],"NGN as a new entry to the specials?")
            confirm = input("Please enter yes or no.\nYour entry: ").lower()                            # Ensuring lowercase input to deactivate case sensitivity
            if confirm == "yes":                                                                 # Condition for a confirmed update 
                special_options.update(spec)                                                           # Now updating specials
                print("**Done. Return to the general operations menu to see the updates.")
                break 
            elif confirm == "no":                                                                # Condition for a no input
                print("Ok, request cancelled! Nothing was added to the specials menu.")        # An output response   
                break                                                                           # End the program by breaking the while loop
            elif confirm != 'yes' or 'no':                                                       # Condition for an input that is neither yes nor no
                print("Invalid entry.")
        else:
            print("Invalid entry for the name and/or price. Start over.")
            break

#FUNCTION REMOVE FROM THE MAIN_MENU
def del_specials_menu():                                               
    while special_options != {}:                                              #Condition for an unempty main menu
        modify = input("""\nWhat meal do you want delete? Please enter the meal name correctly.\nOr press Enter to exit.
        Your entry:""").lower()                                                        # Ensuring lowercase input to deactivate case sensitivity
        if modify != "" and modify in special_options:                                  #Condition for given input that's on the specials menu
            print("Do you want to remove", modify.upper(), "from your specials?")
            while True:
                confirm = input("Please enter yes or no.\nYour entry: ").lower()                            # Ensuring lowercase input to deactivate case sensitivity
                if confirm == "yes":                                                       # Condition for a confirmed update 
                    special_options.pop(modify)
                    print("**Entry deleted. Return to the general operations menu to see the updates.")
                    break 
                elif confirm == "no":                                                                # Condition for a no input
                    print("Okay, request cancelled!", modify.upper(), "was not removed from the specials")        # An output response   
                    break                                                                           # End the program by breaking the while loop
                elif confirm == "":
                    print("No input given.")
                else:                                                                   # Condition for an input that is neither yes nor no
                    print("Invalid Entry.")
                    break
        elif modify == "":
            print("Ok, no changes made.")
            break
        else:
            print("""Invalid Entry. Only a meal name on the menu is allowed.
    If you entered a meal name, check that you entered it correctly then retry.""")
    if special_options == {}:                                                        #Condition for an empty specials menu
        print("This menu is currently unavailable. No changes can be made.")
#FUNCTION TO EDIT THE NAME OF A MEAL ON EITHER MENU
def meal_name_edit(x):
    while x != {}:        
        meal_name = input("""What meal do you want change from the menu? Please enter the meal name correctly.\nOr press Enter to exit.
        Your entry:""").lower()                                                                     # Ensuring lowercase input to deactivate case sensitivity
        if meal_name != "" and meal_name in x:
            new_name = input("Enter the new name to be assigned to it:")
            while new_name != "":                
                print("Do you want to change", meal_name.upper(), "to", new_name.upper(), "on the menu?")
                confirm = input("Please enter yes or no to confirm request.\nYour entry: ").lower()       # Ensuring lowercase input to deactivate case sensitivity
                if confirm == "yes":                                                                 # Condition for a confirmed update 
                    x[new_name] = x.pop(meal_name)                                                   # Replacing the old name with the new
                    print("""**Done. Return to the General Operations Menu to see the updates.
                    The new name will be at the bottom of the list.""")
                    break 
                elif confirm == "no":                                                                 # Condition for a no input
                    print("Ok, request cancelled! No changes made.")                                  # An output response   
                    break                                                                            # End the program by breaking the while loop
                elif confirm == "":
                    print("No input given.")
                else:                                                                     # Condition for an input that is neither yes nor no
                    print("Invalid Entry.")
                    break                    
            else:
                print("Invalid entry for new name. Please restart request.")        
        elif meal_name == "":
            print("Ok, no changes made.")
            break
        else:
            print("Entry error. That is not a valid meal option.\nPlease check your entry spelling then retry.")
            continue

    if x == {}:
        print("The menu you've selected is unavailable. No changes can be made.")          
    return

def new_meal_price():
    """This function asks for the new meal price, takes the user input and returns it as an integer.
    It gives a message for non-integer input."""
    try:
        choice = input("Enter the new price to be assigned to it.\nYour Entry is:")
        return int(choice)
    except ValueError:
        print("Price entry error.\nOnly numbers are valid. Please check your entry and retry.")
        
        
#FUNCTION TO EDIT THE PRICE OF A MEAL ON EITHER MENU
def meal_price_edit(x):
    while x != {}:        
        meal_name = input("""What meal do you want change it's price on the menu? Please enter the meal name correctly.\nOr press Enter to exit.
        Your entry:""").lower()     # Ensuring lowercase input to deactivate case sensitivity
        if meal_name != "" and meal_name in x:
            print(meal_name.upper(), "is currently", x[meal_name],"NGN.")                       
            y = new_meal_price()                
            while y != "" and y !=None:                
                print("Do you want to change", x[meal_name],"NGN to", y,"NGN as the price of", meal_name.upper(),"on the menu?")
                confirm = input("Please enter yes or no to confirm request.\nYour entry: ").lower()                            # Ensuring lowercase input to deactivate case sensitivity
                if confirm == "yes":                                                                 # Condition for a confirmed update 
                    x[meal_name] = y                                                                 # Replacing the value name with the new
                    print("**Done. Return to the General Operations Menu to see the updates.")
                    break 
                elif confirm == "no":                                                                 # Condition for a no input
                    print("Ok, request cancelled! No changes made.")                                  # An output response   
                    break                                                                            # End the program by breaking the while loop
                elif confirm == "":
                    print("No input given.")
                else:                                                                     # Condition for an input that is neither yes nor no
                    print("Invalid Entry.")
                    break                    
            else:
                print("Invalid entry for new price. Please restart request.")        
        elif meal_name == "":
            print("Ok, no changes made.")
            break
        else:
            print("Entry error. That is not a valid meal option.\nPlease check your entry spelling then retry.")
            continue

    if x == {}:
        print("The menu you've selected is unavailable. No changes can be made.")          
    return
        
#FUNCTION TO PRINT MANAGER OPERATIONS AND TAKE MANAGER'S SELECTED OPERATION
def admin_ops():
    """This function gives instructions for the manager input, takes the manager input and returns it as an integer.
    It gives a message for non-integer input."""
    try:
        choice = input("""\nPlease enter:
        [1] To delete an entry from the existing customer order.  
        [2] To add a meal to the main menu.
        [3] To remove a meal from the main menu.
        [4] To add a meal to the specials menu.
        [5] To remove a meal from the specials menu. 
        [6] To change the NAME of a meal on the MAIN MENU
        [7] To change the NAME of a meal on the SPECIALS MENU
        [8] To change the PRICE of a meal on the MAIN MENU
        [9] To change the PRICE of a meal on the SPECIALS MENU
        [10] To restart/exit the Manager Menu.\nYour entry:""")
        return int(choice)                                                             #Ensuring the user input is a number
    except ValueError:
        print("Process selection error.\nOnly numbers are valid. Please check your entry and retry.")
        

#TO CALL FOR FUNCTIONS THAT CONTROL THE MANAGER'S SELECTED OPERATION
def manager():  
    while True:
        next = admin_ops()
        if next == 1:
            print("\n**You're about to delete a meal entry from the CUSTOMER'S ORDER.")
            change()
            continue
        elif next == 2:
            print("\n**You're about to add a meal entry to the MAIN MENU.")
            main_update()
            continue                          
        elif next == 3:
            print("\n**You're about to delete a meal entry from the MAIN MENU.")
            del_main_menu()
            continue
        elif next == 4:
            print("\n**You're about to add a meal entry to the SPECIALS MENU.")
            specials_update()
            continue
        elif next == 5:
            print("\n**You're about to delete a meal entry from the SPECIALS MENU.")
            del_specials_menu()
            continue
        elif next == 6:          
            print("\n**You're about to change the NAME of a meal on the MAIN MENU.")
            x = menu_options
            meal_name_edit(x)
            continue
        elif next == 7:
            print("\n**You're about to change the NAME of a meal on the SPECIALS MENU.")
            x = special_options
            meal_name_edit(x)
            continue
        elif next == 8:
            print("\n**You're about to change the PRICE of a meal on the MAIN MENU.")
            x = menu_options
            meal_price_edit(x)
            continue
        elif next == 9:
            print("\n**You're about to change the PRICE of a meal on the SPECIALS MENU.")
            x = special_options
            meal_price_edit(x)
            continue
        elif next == 10:          
            print("Restarting Manager Mode.")
            break
        else:
            print("Invalid entry. Only 1, 2, 3, 4, 5, 6, and 7 are valid. Please check your entry and retry.")
            continue   
        
#FUNCTION TO PRINT GENERAL OPERATIONS AND TAKE THE USER'S SELECTED OPERATION       
def operation():
    """This function gives instructions for the user input, takes the user input and returns it as an integer.
    It gives a message for non-integer input."""
    try:
        choice = input("""\nThis is our General Operations Menu.
        Please enter:\n[1] To Start/Continue Ordering.\n[2] To See Your Order.\n[3] To Finish and Pay.\n[4] To Get Assistance/Access Manager Mode.\n[5] To exit.
        Your entry:""")
        return int(choice)                                                           
    except ValueError:
        print("Process selection error.\nOnly 1, 2, 3, 4 and 5 are valid options. Please check your entry and retry.")
        
        
#TO CALL FOR FUNCTIONS THAT CONTROL THE USER'S SELECTED OPERATION                  
def proceed():   
    while True:
        next = operation()
        if next == 1:
            see_menu()                    #To see menu options
            continue
        elif next == 2:
            print("You have ordered:")            
            so_far()                     #To see current order
            continue                        
        elif next == 3:
            receipt()                   #To receipt current order
            break
        elif next == 4:
            user()                      #To select the manager mode
            continue
        elif next == 5:
            print("We hate to see you go! Hope to get yor order soon.")       
            break                        #To exit or end the program
        else:
            print("Invalid entry. Only 1, 2, 3, 4 and 5 are valid options. Please check your entry and retry.")
            continue                     #To handle non-operation selection

print("YOU'RE WELCOME TO OKO IYA'LA'MALA KITCHEN!")
proceed()
```
